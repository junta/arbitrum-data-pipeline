```python
# Snapshot Proposal + Forum Topic

import pandas as pd

cleaned_proposals = pd.read_parquet("output_data/snapshot/cleaned_proposals.parquet")
forum_topics = pd.read_parquet("output_data/forum/topics.parquet")

proposal_merged_df = pd.merge(
    cleaned_proposals, forum_topics, left_on="forum_topic_id", right_on="id"
)
print(proposal_merged_df)
```

```python
# Karma Delegates + Snapshot Voting

import pandas as pd

karma_delegates = pd.read_parquet("output_data/karma/delegates.parquet")
voting = pd.read_parquet("output_data/snapshot/votes.parquet")
voting['voter'] = voting['voter'].str.lower()

voting_merged_df = pd.merge(
    karma_delegates, voting, left_on="publicAddress", right_on="voter"
)
print(voting_merged_df)
```
